```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(1, 6)
y = x ** 2
```

# 1️⃣ Line Plot-Stock price trends over time

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

# 2️⃣ Scatter Plot-	Correlation between study hours & marks

```python
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.show()
```

# 3️⃣ Bar Chart-Compare sales of 3 products

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

# 4️⃣ Horizontal Bar Chart- Ranking of players by runs scored

```python
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Chart")
plt.show()
```

# 5️⃣ Histogram- Distribution of exam scores

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='green', edgecolor='black')
plt.title("Histogram")
plt.show()
```

# 6️⃣ Pie Chart- Market share of companies

```python
sizes = [30, 45, 25]
labels = ['A', 'B', 'C']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

# 7️⃣ Stack Plot- Components of electricity usage (AC, fridge)

```python
days = [1,2,3,4,5]
A = [1,2,3,4,5]
B = [1,1,2,2,3]
plt.stackplot(days, A, B, labels=['A','B'])
plt.legend()
plt.title("Stack Plot")
plt.show()
```

# 8️⃣ Area Chart (like fill-between) - 	Rainfall accumulation over time

```python
plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.plot(x, y, color="blue")
plt.title("Area Chart")
plt.show()
```

# 9️⃣ Multiple Lines - 	Compare two stock prices over time

```python
plt.plot(x, y, label='x^2')
plt.plot(x, x*3, label='3x')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

# 🔟 Subplots - Dashboards with multiple plots

```python
fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y)
axs[0].set_title("Plot 1")
axs[1].bar(categories, values)
axs[1].set_title("Plot 2")
plt.show()
```

# **Seaborn**

Import & Data

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')
```

# 1️⃣ Scatter Plot- 	Relationship between income & spending

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Scatter Plot")
plt.show()
```

# 2️⃣ Line Plot - 	Website traffic over a week

```python
sns.lineplot(x='size', y='tip', data=df)
plt.title("Line Plot")
plt.show()
```

# 3️⃣ Bar Plot- Average salary by department

```python
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Bar Plot")
plt.show()
```

# 4️⃣ Count Plot - Number of students per grade

```python
sns.countplot(x='day', data=df)
plt.title("Count Plot")
plt.show()
```

# 5️⃣ Box Plot- Compare salaries across departments

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Box Plot")
plt.show()
```

# 6️⃣ Violin Plot - 	Distribution of house prices by city

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violin Plot")
plt.show()
```

# 7️⃣ Strip Plot- 	Individual customer spending by day

```python
sns.stripplot(x='day', y='total_bill', data=df, jitter=True)
plt.title("Strip Plot")
plt.show()
```

# 8️⃣ Swarm Plot- Individual marks with no overlap

```python
sns.swarmplot(x='day', y='total_bill', data=df)
plt.title("Swarm Plot")
plt.show()
```

# 9️⃣ Heatmap- 	Correlation between variables in sales

```python
corr = df.corr(numeric_only=True)  # add numeric_only=True
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

# 🔟 Pair Plot- 	Explore all variable relationships in dataset

```python
sns.pairplot(df)
plt.show()
```

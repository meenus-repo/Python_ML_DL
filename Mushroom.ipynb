```python
# Confusion Matrix:
```

```python
# To evaluate and summarize the performance of a machine learning classification model 
# by comparing its predicted outputs against the actual, known values in a dataset
```

```python
# To evaluate classification model performance by breaking down predictions into true positives, true negatives, false positives, and false negatives, 
# offering a deeper understanding of model errors and biases than accuracy alone. 
# It enables calculation of key metrics like precision and recall, 
# crucial for balancing the costs of different types of errors in real-world applications such as medical diagnosis or fraud detection. 
```

```python
# Confusion Matrix is used for:

# Detailed Error Analysis (TP, TN, FP, FN)
# Beyond Accuracy - A confusion matrix provides a more granular view of performance. 
# Key Performance Metrics - Precision, Recall(Sensitivity), Specificity, F1 Score.
# Application-Specific Insights - Medical Diagnosis, Fraud Detection.
# Visualizing Performance - Easy to understand how well the model classifies different classes.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_test,y_pred):
  cm=confusion_matrix(y_test,y_pred)
  sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
  plt.title("\n Confusion Matrix\n")
  plt.xticks([0.5,1.5],["Edible","Poisonous"])
  plt.yticks([0.5,1.5],["Edible","Poisonous"])
```

```python
# Decision Tree Classifier
```

```python
# Advantages:
#     Can handle both numerical and categorical data and
#     do not require extensive preprocessing such as Feature scaling or normalization.
#     Decision Trees are robust to outliers in the dataset, 
#     meaning outlier values have minimal impact on overall model performance.
#     They can capture non-linear relationships between input features and the target variable.
#     Decision Trees can handle missing data by making splits based on available information.
#     They are Versatile - supporting both classification and regression problems as well as 
#     multi-output tasks.
#     Trees provide feature importance rankings - aiding in understanding which features drive predictions
#     No need for feature Scaling - They work natively with raw or minimally processed data.
#     Trees have high computationally efficiency and low memory requirements for 
#     small to medium data-sets, supporting real-time usecases.
```

```python
# Disadvantages:
#     Prone to overfitting
#     Unstable - Small changes in training data may result in substantially different trees and predictions.
#     Trees tend to be biased toward dominant features or classes.
#     They can struggle with very large datasets
#     Difficulty in finding complex feature interactions.
```

```python
# Key methods to prevent overfitting in Decision Tree:
#    Pruning the Tree (Remove unimportant|weak branches post|while training)
#    Limiting maximum Depth (Restricts tree complexity)
#    Setting minimum samples for Split|Leaf (Prevents over-specific branches)
#    Limiting the number of Leaf Nodes (Limits overall tree size)
#    Feature selection and Engineering (Focuses on meaningful features, removes irrelevant ones)
#    Ensemble Methods (Averages out individual tree Errors, reduces variance)
#    Cross-Validation (Ensures Model isn't tuned to random data pecularities)
#    Increasing Training Data (Encourages learning of broader data patterns)
#    Dimensionality reduction (Simplifies the task, especially with high feature counts)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier()
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
print('\n Decision Tree')
print(classification_report(y_test,y_pred,target_names=target_names))
plot_confusion_matrix(y_test,y_pred)
```
